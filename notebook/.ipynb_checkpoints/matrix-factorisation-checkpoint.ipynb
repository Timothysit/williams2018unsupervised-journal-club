```python
import os
import numpy as np
import bokeh.plotting as bkplot
import pandas as pd
import bokeh.io as bkio 
import bokeh.models as bkmodel
bkio.output_notebook()
from collections import defaultdict
import matplotlib.pyplot as plt

import sciplotlib.style as splstyle
import sciplotlib.polish as splpolish
```

```python
# make random matrix

num_neuron = 10
num_time_points = 25

X = np.random.normal(0, 2, (num_neuron, num_time_points))


# Actually get some patterned activity 
# maybe a sine curve plus a linear line?

constant_scaling = 5
noise_std = 0.05

sine_latent = np.sin(np.linspace(-np.pi, np.pi, num_time_points)) * constant_scaling
increase_latent = np.linspace(1, 2, num_time_points) * constant_scaling

neuron_activity_list = list()
for neuron in np.arange(num_neuron):
    
    sine_latent_weight = np.random.uniform(0, 1)
    increase_latent_weight = 1 - sine_latent_weight
    # increase_latent_weight = np.random.uniform(0, 1) 
    noise = np.random.normal(0, noise_std, (num_time_points)) * constant_scaling
    neuron_activity = sine_latent * sine_latent_weight + increase_latent * increase_latent_weight + noise 
    neuron_activity_list.append(neuron_activity)

X = np.vstack(neuron_activity_list)
```

```python
import scipy.stats as sstats

with plt.style.context(splstyle.get_style('nature-reviews')):
    fig, ax = plt.subplots()
    a = 1.99
    x = np.linspace(sstats.gamma.ppf(0.0, a),
                    sstats.gamma.ppf(0.99, a), 100)
    input_1 = sstats.gamma.pdf(x, a)
    input_2 = sstats.gamma.pdf(x, 5)
    input_3 = np.repeat(0.1, len(x))
    ax.plot(x, input_1, label='Input 1')
    ax.plot(x, input_2, label='Input 2')
    ax.plot(x, input_3, label='Input 3')
    ax.plot(x, input_1 + input_2 + input_3, label='Neural activity')
    
    ax.set_ylabel('Firing rate', size=12)
    ax.set_xlabel('Time', size=12)
    ax.legend()
    
    fig_folder = '/home/timsit/williams2018unsupervised-journal-club/figures/tca-concepts/'
    fig_name = 'activiyt_as_linear_combination_of_inputs'
    fig.savefig(os.path.join(fig_folder, fig_name), dpi=300, bbox_inches='tight')
    # fig, ax = splpolish.set_bounds(fig, ax)
```
